```python
import pandas as pd
%matplotlib inline
import geopandas as gpd
import matplotlib.pyplot as plt
df = pd.read_csv("SidewalkStepsBlockgroup.csv",index_col="Neighborho", dtype={8: str}) 
df = df.dropna()
df.head(10)
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# Plot only the top 10 neighborhoods by length of sidewalk
df_filtered = df[df['LengthMile'] < 1000] 
df_top10 = df_filtered.groupby('Neighborho')['LengthMile'].sum().nlargest(25).reset_index()
df_top10.plot(kind='bar', x='Neighborho', y='LengthMile', legend=False)
plt.title('Top 25 Neighborhoods by Sidewalk Length')
plt.xlabel('Neighborhood')
plt.ylabel('Sidewalk Length (in miles)')
plt.show()
```

```python
df = pd.read_csv("SidewalkStepsBlockgroup.csv",index_col="Neighborho", dtype={8: str}) 
num_miles = df.groupby("Neighborho").sum()['LengthMile']
num_miles.sort_values(ascending=False)
```

```python
neighborhoods = gpd.read_file("Neighborhoods_.shp")
neighborhoods.plot() 
```

```python
sidewalk_map = neighborhoods.merge(num_miles, how='left', left_on='hood', right_on='Neighborho')
#sidewalk_map = sidewalk_map.dropna()
sidewalk_map[['hood','LengthMile']].head(10)
```

```python
sidewalk_map.plot(column='LengthMile')
```

```python
sidewalk_map.plot(column='LengthMile', 
               cmap='OrRd',             
               edgecolor="white",        
               legend=True,              
               legend_kwds={'label': "Number of Miles of Sidwalk"}, 
               figsize=(15, 10),        
               missing_kwds={"color": "lightgrey"} 
               )
```
