# Parks, by Ava 

## Dataset Introduction
In this project we are defining the "best" neighborhood in Pittsburgh as the one which is gives the most opportunity for its members to be healthy. For this dataset, we are analyzing the number of public parks in each neighborhood, with the assumption that having more parks in a neighborhood has a positive impact on the health of the residents. This is because having an open park would encourage more residents to go outside and exercise.  

* First, I loaded the dataset as a csv and created a datatable. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

parks = pd.read_csv("parks.csv")
parks.head()
```

* Then, I created a dictionary based off of neighborhood to count the total number of parks in each neighborhood

```python
counter = dict()
for row in parks["neighborhood"]:
    
    if row not in counter:
        counter[row] = 1
    else:
        counter[row] = counter[row] + 1

print(counter)
len(counter)
```

* I then displayed the data as a bar graph

```python
df = pd.DataFrame(list(counter.items()), columns=['Neighborhood', 'Number of Parks'])

df.plot(kind='bar', x='Neighborhood', y='Number of Parks', legend=False)
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.title('Bar Graph')
plt.show()
```

* To make the bar graph easier to read, I sorted it in decreasing order and adjusted the margins

```python
df = pd.DataFrame(list(counter.items()), columns=['Neighborhood', 'Number of Parks']).sort_values(by='Number of Parks', ascending=False)
plt.figure(figsize=(30, 15)) 
ax = df.plot(kind='bar', x='Neighborhood', y='Number of Parks', legend=False, ax=plt.gca()) 

plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks', fontsize = 50)
plt.title('Number of Parks by Neighborhood', fontsize = 50)
plt.xticks(rotation=45, ha="right")
plt.gca().margins(x=0.02)
plt.tight_layout()

plt.show()
```

* Then, to develop an area heat map, I imported geopandas and used it to create a datatable

```python
import geopandas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

parks = pd.read_csv("parks.csv")
num_parks = parks.groupby("neighborhood")['_id'].count()
num_parks.sort_values(ascending=False)
print(num_parks)
```

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
parks_map = neighborhoods.merge(num_parks, how='left', left_on='hood', right_on='neighborhood')
parks_map[['hood','_id','geometry']].head()
parks_map.plot(column='_id', 
               cmap='OrRd',  
               edgecolor="black",   
               legend=True, 
               legend_kwds={'label': "Number of Parks"}, 
               figsize=(15, 10),   
               missing_kwds={"color": "lightgrey"}
               )
```

# Courts, by Kolja

## Introduction
An important part of you physical and mental health is playing the sports you love. If there are no sports courts around you, you are forced to either run or do home workout routines, which aren't as fun and have detrimental impact on your mental health which also impacts your physical health. That's why having more courts in your neighborhood is a metric that solidifies your neighborhood as the best.

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

First I imported the data (list with all of the courts)

```python
courts = pd.read_csv("list-Copy1.csv")
courts.head(5)
```

Then I organized it a bit and put the data into a dictionary with the neighborhood accompanied by the number of courts in it.

```python
court_count = {} 
quantity = 1

for index, row in courts.iterrows():
    item_name = row['neighborhood']
    if item_name in court_count:
        court_count[item_name] += quantity
    else:
        court_count[item_name] = quantity

print(court_count)
```

I created a dataframe for it

```python
# Convert the dictionary to a DataFrame
court_counts_df = pd.DataFrame(list(court_count.items()), columns=['neighborhood', 'Court_Count'])
```

I followed it by making a bar chart of top 25 neighborhoods with the most courts, which would help me see how the best stack up.

```python
df_top = court_counts_df.groupby('neighborhood')['Court_Count'].sum().nlargest(25).reset_index()

df_top.plot(kind='bar', x='neighborhood', y='Court_Count', legend=False)

plt.title('Courts in Pittsburgh')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Courts')
plt.show()
```

I then loaded the map

```python
# Load the neighborhoods shapefile
neighborhoods = gpd.read_file("Neighborhoods/Neighborhoods_.shp")
neighborhoods.plot() # plot the geopandas dataframe
plt.show()
```

Lastly I added my data into the map and organised its colours to better show which neigborhood has the most courts.

```python
# Merge the court counts DataFrame with the neighborhood shapefile data
courts_map = neighborhoods.merge(court_counts_df, how='left', left_on='hood', right_on='neighborhood')
```

```python
courts_map[['hood','Court_Count','geometry']].head()
```

```python
courts_map.plot(column='Court_Count')
plt.show()
```

```python
courts_map.plot(column='Court_Count', # set the data to be used for coloring
               cmap='YlGnBu',              # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Court_Count"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "white"} # set disctricts with no data to gray
               )
```
